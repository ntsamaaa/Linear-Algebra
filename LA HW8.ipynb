```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=np.linspace(-1,1, 100)
b=np.linspace(-1,1, 100)
plt.axvline(0, c='gray')
plt.axhline(0, c='gray')
plt.scatter(a, b, s=10)
```

```python
OriginalLine=np.array([a, b]).T
```

NewLine =A + tran@OriginalLine

```python
A=np.array([3, 4])
tran=np.array([[2, -3], [-4, 7]])
plt.axvline(0, c='gray')
plt.axhline(0, c='gray')
for i in range(OriginalLine.shape[0]):
    tmp = A + np.dot(tran, OriginalLine[i])
    plt.scatter(tmp[0], tmp[1], s = 10, c = 'y')
```
